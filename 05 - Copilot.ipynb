```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='dataType')
plt.xlabel('Data Type')
plt.ylabel('Count')
plt.title('Count of Each Data Type')
plt.show()
```

```python
import matplotlib.pyplot as plt
df['dataType'].value_counts().plot(kind='bar')
plt.xlabel('Data Type')
plt.ylabel('Count')
plt.title('Count of Each Data Type')
plt.show()
```
