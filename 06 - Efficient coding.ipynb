# Leverage Configuration Parameters with Defaults

When writing functions, always provide default parameters so that functions can run with minimal inputs while still allowing customization. This will make your code cleaner and easier to use, especially when working in notebooks where you might want to quickly test something with default settings.

```python
def process_data(df, dropna=True, fillna_value=None, partition_col=None):
    if dropna:
        df = df.dropna()
    if fillna_value:
        df = df.fillna(fillna_value)
    if partition_col:
        df = df.repartition(partition_col)
    return df
```

```python
# Usage:
df_processed = process_data(df, fillna_value=0, partition_col="Age")
df_processed.show()
```

# Use Decorators for Reusable Logic

```python
import time
```

```python
def timeit(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        print(f"Execution time: {time.time() - start_time} seconds")
        return result
    return wrapper
```

```python
@timeit
def run_transformation(df):
    return df.filter(df["age"] > 18)
```

```python
df = spark.sql("SELECT * FROM Lakehouse.dbo.fabcon01 LIMIT 1000")
df_tra = run_transformation(df)
```
