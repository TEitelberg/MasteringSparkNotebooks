```python
import pandas as pd
from faker import Faker
import ipywidgets as widgets
from IPython.display import display
from pyspark.sql import SparkSession
from pyspark.sql import Row
import ipywidgets as widgets
from IPython.display import display
import time
```

```python
# Globale Variablen
selected_checkboxes = {}
slider_value = None
entered_name = None
generated_df = None
```

```python
# Initialisiere die Faker-Bibliothek
fake = Faker()
```

```python
# Funktion zum Erstellen eines Formulars
def create_form():
    global selected_checkboxes, slider_value, entered_name
    
    # Textbox für die Namenseingabe
    name_input = widgets.Text(
        value='',
        placeholder='Enter your name',
        description='Name:',
        disabled=False
    )
    
    # Slider von 100 bis 1.000.000
    value_slider = widgets.IntSlider(
        value=100,
        min=100,
        max=1000000,
        step=100,
        description='Value:',
        continuous_update=True
    )
    
    # Checkboxen
    check_firstname = widgets.Checkbox(value=False, description='Firstname')
    check_lastname = widgets.Checkbox(value=False, description='Lastname')
    check_age = widgets.Checkbox(value=False, description='Age')
    check_gender = widgets.Checkbox(value=False, description='Gender')
    check_country = widgets.Checkbox(value=False, description='Country')
    check_email = widgets.Checkbox(value=False, description='Email')

    # Schaltfläche "Generate"
    generate_button = widgets.Button(
        description='Generate',
        button_style='success',  # Green button
        tooltip='Click to generate name and value',
        icon='check'
    )
    
    # Funktion, die ausgeführt wird, wenn die Schaltfläche gedrückt wird
    def on_button_click(b):
        global selected_checkboxes, slider_value, entered_name, generated_df
        
        # Markierte Checkboxen
        selected_checkboxes = {
            'Firstname': check_firstname.value,
            'Lastname': check_lastname.value,
            'Age': check_age.value,
            'Gender': check_gender.value,
            'Country': check_country.value,
            'Email': check_email.value
        }
        
        # Wert des Sliders
        slider_value = value_slider.value
        
        # Inhalt der Textbox
        entered_name = name_input.value
        
        # Erstelle das DataFrame basierend auf den ausgewählten Checkboxen
        generated_df = create_dataframe(selected_checkboxes, slider_value)
        write_deltatable(entered_name, generated_df)
        # print(df)

        # Ausgabe
        print("Selected checkboxes:", selected_checkboxes)
        print("Slider value:", slider_value)
        print("Entered name:", entered_name)
        print(f"Table {entered_name} was created with {slider_value} rows.")

    # Verknüpfe die Schaltfläche mit der Funktion
    generate_button.on_click(on_button_click)
    
    # Zeige die Widgets an
    display(widgets.VBox([
        name_input,
        value_slider,
        check_firstname, check_lastname, check_age, check_gender, check_country, check_email,
        generate_button
    ]))
```

```python
def create_dataframe(selected_fields, num_rows):
    data = []

    # Generiere die Datenzeilen basierend auf den markierten Checkboxen
    for i in range(num_rows):
        row_data = {}
        if selected_fields['Firstname']:
            row_data['Firstname'] = fake.first_name()
        if selected_fields['Lastname']:
            row_data['Lastname'] = fake.last_name()
        if selected_fields['Age']:
            row_data['Age'] = fake.random_int(min=18, max=90)
        if selected_fields['Gender']:
            row_data['Gender'] = fake.random_element(elements=('Male', 'Female', 'Other'))
        if selected_fields['Country']:
            row_data['Country'] = fake.country()
        if selected_fields['Email']:
            row_data['Email'] = fake.email()
        data.append(Row(**row_data))

    # Erstelle ein Spark DataFrame aus den Daten
    df = spark.createDataFrame(data)
    return df
```

```python
def write_deltatable(tablename, df_source):

    spark.sql(f"DROP TABLE IF EXISTS dbo.{tablename}")
    
    df_source.write\
        .format("delta")\
        .mode("overwrite")\
        .option("parquet.vorder.enabled ","true")\
        .saveAsTable(f"dbo.{tablename}")
```
