```python
from pyspark.sql.functions import col, sum as spark_sum
import random


data = [(random.randint(1, 100), random.randint(1, 100), random.randint(1, 100), random.randint(1, 100)) for _ in range(3)]

df = spark.createDataFrame(data, ["Spalte1", "Spalte2", "Spalte3", "Spalte4"])
df = df.withColumn("Summe", col("Spalte1") + col("Spalte2") + col("Spalte3") + col("Spalte4"))


display(df)
```

```python
df.createOrReplaceTempView("random_numbers")
```

```python
%%pyspark

result = spark.sql("SELECT * FROM random_numbers")

display(result)
```
